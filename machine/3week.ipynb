
# **3주차_Percentron 기초**

## (wx1 + wx2) x b  xnor gate에 따라서 여러가지 것들이 존재

##  h(x) = 1/ (1+ e^x)  // h'(x)= h(x) x (1-h(x))

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/ (1+ np.exp(-x))
x= np.array([-1,1,2])
y= sigmoid(x)
plt.plot(x,y)
plt.ylim(0,1.1)
plt.show()
```

```python
def relu(x):
    return np.maximum(0,x)
x=np.arange(-6,6,0.1)
y=relu(x)
plt.plot(x,y)
plt.hlines(0,-6,6)
plt.vlines(0,0,6)
plt.show()
```

# **3주차_신경망 기초**

##### w b 즉 가중치와 편향값을 수정해서 목표값으로 도출 해내는 기술

```python
x=2
t=10
w=3
b=1
lr=0.1
for i in range(30):
    y=x * w + b
    print("y : %6.3f" %y)
    yb= y-t ## 나의 목표값 - 추정값
    xb =yb*w
    wb= yb*x
    bb=yb * 1
    print("x = %6.3f , w= %6.3f , b= %6.3f " % (xb,wb,bb))
    w= w-(lr*wb)
    b= b-(lr*bb)
    print("x = %6.3f , w= %6.3f , b= %6.3f " % (x,w,b))

```

##### 출력값에서 바라볼때 w 1 w 2 형식으로 선정하게 된다! 넘버링을 할 당시

```python
x1, x2 = 2, 3
w1, w2 = 3, 4
w3,w4 = 4,6
t1,t2 = 27, -3
b1,b2 = 1, 2
lr = 0.01  # 학습률

for i in range(200):
    #수정필요
    y1 = x1 * w1 + x2 * w2 + b1
    y2 = x1 * w3 + x2 * w4 + b2
    print("%i 번째 : y1 = %6.3f / y2 = %6.3f " % (i + 1, y1,y2))

    #수정해야함
    E = ((y1 - t1) ** 2 / 2)+((y2 - t2) ** 2 / 2)
    print("E = %6.3f " % E)
    
    # 손실이 임계값보다 작을 때 학습 종료
    if E < 0.000001:
        break
    
    # 기울기 계산
    dw11 = (y1 - t1) * x1
    dw21 = (y1 - t1) * x2
    db1 = (y1 - t1)
    dw12 = (y2 - t2) * x1
    dw22 = (y2 - t2) * x2
    db2 = (y2 - t2)
    
    # 가중치 및 편향 업데이트
    w1 = w1 - lr * dw11
    w2 = w2 - lr * dw21
    b1 = b1 - lr * db1
    
    w3 = w3 - lr * dw12
    w4 = w4 - lr * dw22
    b2 = b2 - lr * db2
    
    

    print("w1 = %6.3f / w2 = %6.3f / b1 = %6.3f" % (w1, w2, b1))
    print("w3 = %6.3f / w4 = %6.3f / b2 = %6.3f" % (w3, w4, b2))

```
